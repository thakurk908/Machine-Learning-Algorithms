```python
# pip install mlxtend
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel("e:\Online Retail.xlsx")
df.head()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
basket = (df[df['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket.head(10)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
# With the lift value, we can interpret the importance of a rule. It is a measure for a rule.
# The lift value of an association rule is the ratio of the confidence of the rule and the expected confidence of the rule.
rules.head()
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
```
